2. Create numpy arrays that have zero, one, two, three, Four dimensions respectively.
Also, print the shape and dimensions of the numpy arrays created above.

```python
# 0-D array (scalar)
arr0 = np.array(42)
print("0-D Array:", arr0)
print("Shape:", arr0.shape, "Dimensions:", arr0.ndim, "\n")

# 1-D array (vector)
arr1 = np.array([1, 2, 3, 4, 5])
print("1-D Array:", arr1)
print("Shape:", arr1.shape, "Dimensions:", arr1.ndim, "\n")

# 2-D array (matrix)
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("2-D Array:\n", arr2)
print("Shape:", arr2.shape, "Dimensions:", arr2.ndim, "\n")

# 3-D array (cube-like)
arr3 = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("3-D Array:\n", arr3)
print("Shape:", arr3.shape, "Dimensions:", arr3.ndim, "\n")

# 4-D array
arr4 = np.array([[[[1, 2], [3, 4]]]])
print("4-D Array:\n", arr4)
print("Shape:", arr4.shape, "Dimensions:", arr4.ndim, "\n")

#Number of dimensions can be decided by seeing their [] brackets
```

9. Create a random array, with the range 0-1000, with a size 50 . Perform the
following operations on the same.
a. Print the first ten elements of the array.
b. Print the last ten elements of the array
c. Print the elements from the indexes 10-25
d. Print the element at the index 22
e. Print the array using negative indexing.
f. Print the last ten elements of the array, using negative indexing.


```python
rand = np.random.randint(0,1001, size = 50)
print("Full array:\n",rand, "\n")

#a. First 10 elements
print("First 10 elements:", rand[:10])

#b. Last 10 elements
print("Last 10 elements:", rand[-10:])

#c
print("Elements from index 10 to 25:", rand[10:26])

#d
print("Ele in index 22:", rand[22],"\n")

#e. Array using negative indexing (reverses the array)
print("Array reversed:\n", rand[::-1], "\n")

#f
print("last 10 elements using neg index:", rand[-10:])
```

10. Given a numpy array that has values = [1,2,3,4,5]. Perform the following operations.
a. Insert the element 6 at the end of the array.
b. Insert the element 10 at the beginning of the array.
c. Insert the element 30, before 4.
d. Update the element 5 as 50.
e. Delete the element 50 in the updated array.
f. Sort the array in ascending order.
g. Sort the array in descending order.


```python
arr = np.array([1,2,3,4,5])
print(arr)

arr = np.append(arr,6)
print("After inserting 6 at end:", arr)

arr=np.insert(arr,0,10)
print("After Inserting the ele 10 at the beg:",arr)

arr=np.insert(arr,4,30)
print(arr)

arr[arr == 5] = 50
print("After updating 5 as 50:", arr)

arr = arr[arr != 50]
print("After deleting 50:", arr)

arr = np.sort(arr)
print("Sorted arr:",arr)


arr = np.sort(arr)[::-1]
print("Sorted arr in descending:",arr)
```

Create a **1-D NumPy array of 20 random integers** between **10 and 100**.  
Print the following:
* Mean  
* Median  
* Standard Deviation  
* Variance  


```python
x = np.random.randint(10, 101, size=20)
print("Random Numbers:", x)
mean=np.mean(x)
print("Mean:",mean)
median=np.median(x)
print("median:",median)
std=np.std(x)
print("Standard deviation:",std)
var=np.var(x)
print("Variance:",var)
```
