```python
# Example forc flatten and ravel
# Create a 2D array
arr1 = np.array([[1, 2, 3],
                [4, 5, 6]])

# Flatten → always returns a copy
flat = arr1.flatten()
print("Flatten result:", flat)

# Ravel → returns a view when possible
rav = arr1.ravel()
print("Ravel result:", rav)

# Modify the flatten output
flat[1] = 77
print("After modifying flatten:\n", arr1)

# Modify the ravel output
rav[0] = 99
print("After modifying ravel:\n", arr1)
```

**8. Array Operators**

**Joining arrays**

```python
A1D = np.array([11,22,33])
B1D = np.array([44,55,66])
```

```python
A1D,B1D
```

```python
np.concatenate((A1D,B1D)) #joins along an existing axis
```

```python
np.vstack((A1D,B1D)) #stacks vertically (row-wise)
```

```python
np.hstack((A1D,B1D)) #stacks horizontally (column-wise)
```

```python
A2D = np.array([[1,2,3,4]])
B2D = np.array([[4,5,6,7]])

print("concatenate axis=0:\n", np.concatenate((A2D,B2D), axis=0))

print("\nconcatenate axis=1:\n", np.concatenate((A2D,B2D), axis=1))


print("\nvstack:\n", np.vstack((A2D,B2D)))

print("\nhstack:\n", np.hstack((A2D,B2D)))
```

**11. Searching Functions**

```python
#np.where() - Returns indices (or values) where a condition is true.

arr3 = np.array([10, 20, 30, 0, 40])
indices = np.where(arr3 > 25)
print(indices)

values = np.where(arr3 > 25, arr3, -1)
print(values) 
```

```python
#Returns the index of the maximum value in the array
np.argmax(arr3)
```

```python
#Returns the index of the minimum value in the array
np.argmin(arr3)
```

```python
#Returns indices of all non-zero elements. It’s similar to np.where(arr != 0)
np.nonzero(arr3)
```

```python
np.sort(arr3)
```

```python
np.unique(arr3)
```

```python
np.argsort(arr3) #Returns indices that sort the array
```
