3.Create a pandas series, with the elements as numbers from 1-10. Use the index values
starting from 0-10. For the above series:
a. Convert it to a list.
b. Convert it to a dictionary.
c. Convert it to a tuple.
d. Convert it to a set.

```python
series = pd.Series(range(1,11),index =range(0,10))
print(series)

l = series.tolist()
print("List:",l)

d = series.to_dict()
print("dictionary:",d)

t = tuple(series)
print("tuple:",t)

s = set(series)
print("set:",s)
```

```python
# Create Series with values 110 and index 010
data = list(range(1, 11))          # 1 to 10
index = list(range(0, 10))         # 0 to 9 (10 indexes)

series = pd.Series(data, index=index)

print("Original Series:\n", series, "\n")

# a. Convert to list
as_list = series.tolist()
print("As List:", as_list)

# b. Convert to dictionary
as_dict = series.to_dict()
print("As Dictionary:", as_dict)

# c. Convert to tuple
as_tuple = tuple(series.tolist())
print("As Tuple:", as_tuple)

# d. Convert to set
as_set = set(series.tolist())
print("As Set:", as_set)
```

4. Create a sample pandas series with x integer elements. Perform the following operations
on the same:
a. Print the first 5 elements of the series
b. Print the last 5 elements of the series
c. Insert an integer value 50 at the index 5.
d. Update the last element in the series with the value 100.
e. Delete the first element from the series.
Note: The index values must not be affected by the operations and should be uniform for
the series after the last operation.

```python
data = list(range(1, 11))  # 110
series = pd.Series(data)

print("Original Series:\n", series, "\n")

# a. Print the first 5 elements
print("First 5 elements:\n", series.head(5), "\n")

# b. Print the last 5 elements
print("Last 5 elements:\n", series.tail(5), "\n")

# c. Insert an integer value 50 at the index 5
series.loc[5] = 50
print("After inserting 50 at index 5:\n", series, "\n")

# d. Update the last element in the series with the value 100
series.iloc[-1] = 100
print("After updating last element to 100:\n", series, "\n")

# e. Delete the first element from the series
series = series.drop(series.index[0])
series = series.reset_index(drop=True)  # reset index to keep uniformity
print("After deleting first element:\n", series, "\n")
```

9. Create a pandas Dataframe with two columns = [Name, Age]. Perform the following
operations on the same.
a. Print the first 3 rows of the Data Frame .
b. Print the rows that have the age greater than 15.
c. Insert 5 imaginary names and ages as values to the pandas Data Frame .
d. Update the 5th row so that the name = john and age=56.
e. Delete the first row from the Data Frame .

```python
df = pd.DataFrame({
    'Name': ['Abhi', 'Bob', 'Cherry', 'Donlee', 'Ephreim'],
    'Age': [10, 20, 14, 18, 12]
})

print("Original DataFrame:\n", df, "\n")

# a. Print the first 3 rows
print("First 3 rows:\n", df.head(3), "\n")

# b. Print rows with age > 15
print("Rows where Age > 15:\n", df[df['Age'] > 15], "\n")

# c. Insert 5 imaginary names and ages
new_data = pd.DataFrame({
    'Name': ['F', 'G', 'H', 'I', 'J'],
    'Age': [25, 30, 17, 40, 22]
})
df = pd.concat([df, new_data], ignore_index=True)
print("After inserting 5 new rows:\n", df, "\n")

# d. Update the 5th row so that name = John and age = 56
df.loc[4, 'Name'] = 'John'
df.loc[4, 'Age'] = 56
print("After updating 5th row:\n", df, "\n")

# e. Delete the first row
df = df.drop(0).reset_index(drop=True)
print("After deleting first row:\n", df)
```

12. Create a Data Frame with columns Name and Age. Perform the following operations.
a. Write a sample code to search for the Name/Names where Age is 20.
b. Sort the Data Frame in ascending order of ages.
c. Using Python loops, print the ages from the data frame and store it in a set.

```python
df = pd.DataFrame({
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Age': [10, 20, 14, 20, 25]
})

print("Original DataFrame:\n", df, "\n")

# a. Search for the Name(s) where Age is 20
names_with_20 = df[df['Age'] == 20]['Name']
print("Names with Age 20:\n", names_with_20.tolist(), "\n")

# b. Sort the DataFrame in ascending order of Age
sorted_df = df.sort_values(by='Age')
print("DataFrame sorted by Age:\n", sorted_df, "\n")

# c. Using Python loops, print ages and store in a set
age_set = set()
for age in df['Age']:
    print("Age:", age)
    age_set.add(age)

print("\nUnique Ages stored in set:", age_set)
```
